```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[5000]])
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]),color='blue')
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = mx+c, where m = coef & c = intercept & area = random variable
167.30954677*3600+76692.3818707813
```

```python
d = pd.read_csv('areas.csv') 
d.head(3)
```

```python
p = reg.predict(d[['area']])
d['prices'] = p
```

```python
d.to_csv('predicted.csv', index=False)
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita(US$)')
plt.scatter(df.year, df.per_capita, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita)
```

```python
reg.predict([[2023]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita(US$)')
plt.scatter(df.year, df.per_capita, color='red', marker='+')
plt.plot(df.year, reg.predict(df[['year']]),color='blue')
```

```python
p = reg.predict([[2025]])
```

```python
d = pd.read_csv('year.csv')
d
```

```python
p = reg.predict(d[['year']])
d['predicted_income'] = p
```
